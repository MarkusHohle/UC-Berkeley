**0) Loading libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

**1) Reading the data set**

```python
XYZ = pd.read_csv("Rot.txt", delim_whitespace = True, header = None)
XYZ = np.array(XYZ)
```

**2) Plotting the data set**

```python
# standard3D plot
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')
ax.scatter(XYZ[:,0], XYZ[:,1], XYZ[:,2], c = 'black', marker = 'o', s = 40)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
plt.show()
```

**3) Running PCA**

```python
out = PCA(n_components = 3).fit(XYZ)

eigenVec = out.components_
eigenVal = out.explained_variance_
eigenXYZ = out.transform(XYZ)
```

**4) Plotting Eigenvalue Spectrum and Data in Eigenspace**

```python
xplot = np.arange(1,4)

plt.bar(xplot, eigenVal, color = (0.8, 0.8, 0.8), edgecolor = 'black')
plt.xlabel('dimension')
plt.ylabel('eigenvalue')
plt.yscale('log')
plt.xticks(xplot)
plt.show()
```

```python
# standard3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')
ax.scatter(eigenXYZ[:,0], eigenXYZ[:,1], eigenXYZ[:,2], c = 'black', marker = 'o', s = 40)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.tick_params(axis = 'both', which = 'major', labelsize = 5)
plt.show()
```
