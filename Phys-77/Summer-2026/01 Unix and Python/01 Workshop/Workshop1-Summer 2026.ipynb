## Exercise 2.3: Printing

Up until this point, we've just been typing a single line of code in each Jupyter cell and running it.  Most Python interpreters will display the result of the final thing you typed, but often you'll want to display the results of many things in a single Python script.

````python
"These are some numbers:"
3*2
3*3
3*4
````

In the cell above, there are several multiplications happening but only the final result is displayed.  To display everything, we simply use a "print statement" on each line.

## Exercise 2.5: Variables

Suppose you calculate something in Python and would like to use the result later in your program (instead of just printing it and immediately throwing it away). One big difference between a calculator and a computer language is an ability to store the values in memory, give that memory block a name, and use the value in later calculations. Such named memory block is called a _variable_. To create a variable, use an _assignment_ opperator =  . Once you have created the variable, you can use it in the calculations.

````python
x = "Phys"
y = "ics!"
z = x + y     # Put 'em together
z             # See what we got!
````

````python
y + x         # Backwards!
````

````python
z[0]          # Grab the first character
````

````python
z[1:3]        # Grab the next two characters
````

````python
z[:4] == x    # Test a match!
````

````python
z*3 + 5*z[-1] # Woo!
````

## Exercise 2.7: Importing libaries

A central part of python is that we don't have to do everything from scratch! If someone else has already made a useful set of functions or objects, we can import these as a library, then use their contents without having to make them ourselves.

Three classic examples of this are the 'math' (used for math), 'numpy' (used for linear algebra).

When we import a library, we can do so under an 'alias', where we use another name for the library. Here, we've imported 'numpy' as 'np'

````python
import math
import numpy as np
````

Once it's been run once we can use these libraries in any cell - often they're imported right at the start!

Here's some examples of what these libraries can do.

````python
print('Cosine 0.5:', math.cos(0.5))
print('Sqrt 9:',math.sqrt(9))
````

````python
print('A lenth 10 array of zeros:', np.zeros(10))
print('11 evenly spaced values between (and including) 0 and 5:',np.linspace(0,5,11))
````
