## 2): Numerical differentiation and Taylor series

Next we'll look at differentiation.

First, plot the function $f(x) = \sqrt{1 + x}$ in the range zero to four. 

```python
x = np.arange(0,4,0.01)
y = np.sqrt(1+x)
```

```python
plt.plot(x,y)
```

Next, derive the 1st and 2nd order Taylor approximation at $x_0 = 0$ and plot the exact function in the same graph together with both approximations. 

```python
yapprox1 = 1 + 0.5*x
yapprox2 = 1 + 0.5*x - 0.125 * x**2

plt.plot(x, y, label = r'$\sqrt{1+x}$')
plt.plot(x, yapprox1, label = r'$1 + \frac{x}{2}$')
plt.plot(x, yapprox2, label = r'$1 + \frac{x}{2} - \frac{x^2}{8}$')
plt.legend()
plt.show
```

Estimate $\sqrt{1.5}$, $\sqrt{2}$ and $\sqrt{3}$ from both approximations and compare the values to the exact values.
