## 2) Preparation

We call all the libraries needed:

## 3) Data

The dataset contains 60,000 8bit RGB images (size 32x32 pixel) of the following classes:

````python
class_names = ["airplane","automobile","bird","cat","deer", "dog","frog","horse","ship","truck"]
````

We first call the data set and split it in training and test set.

````python
(X_train, Y_train), (X_test, Y_test) = cifar10.load_data()
````

````python
X_train.shape
````

````python
X_test.shape
````

Since the images are encoded in 8bit, we can easly normalize the data (```scaler``` not needed).

````python
X_train = X_train / 255.0
X_test  = X_test / 255.0

# One-hot encode labels
Y_train = to_categorical(Y_train, 10)
Y_test  = to_categorical(Y_test, 10)
````

Next, we plot a few images in order to get an idea of what they look like:

````python
# Convert labels to 1D if one-hot encoded
if len(Y_train.shape) > 1:
    labels = np.argmax(Y_train, axis=1)
else:
    labels = Y_train.flatten()

# Collect indices for each class
class_indices = {i: np.where(labels == i)[0][:10] for i in range(10)}

plt.figure(figsize = (12, 12))

# Plot: rows = samples, columns = classes
for col in range(10):          # each column = one class
    for row in range(10):      # 10 images per class
        idx = class_indices[col][row]
        
        plt.subplot(10, 10, row * 10 + col + 1)
        plt.imshow(X_train[idx])
        plt.axis('off')
        
        # Set column titles (once at top)
        if row == 0:
            plt.title(class_names[col], fontsize = 8)

plt.tight_layout()
plt.show()
````
