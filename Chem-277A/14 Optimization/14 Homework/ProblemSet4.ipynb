## 1) The Model
We will represent a protein as a set of coordinates: coords = [(x0, y0), (x1, y1), ... , (xN, yN)].

Then, we define our energy as:
$$
E(\{\mathbf{r}_i\}) =
\sum_{i=1}^{n-1} (\|\mathbf{r}_i - \mathbf{r}_{i+1}\| - 1)^2
+
\lambda \sum_{i=1}^{n} \sum_{j=i+2}^{n}
\frac{1}{(\|\mathbf{r}_i - \mathbf{r}_j\| + \epsilon)^6}
$$

The first term is a bond constrain and will keep chain bonds length around 1, the second term will repel all points to prevent overlap, together this will create many local minima.

Given below is a function called `energy` that calculates the energy of a protein given a set of coordinates and a lambda parameter.



```python
def distance(a, b):
    return np.linalg.norm(np.array(a) - np.array(b))


def energy(coords, lam=0.1):
    E = 0.0
    n = len(coords)

    # bond constraint
    for i in range(n - 1):
        r = distance(coords[i], coords[i+1])
        E += (r - 1)**2

    # repulsion
    for i in range(n):
        for j in range(i+2, n):  # skip neighbors
            r = distance(coords[i], coords[j]) + 1e-6
            E += lam * (1 / r**6)

    return E
```
