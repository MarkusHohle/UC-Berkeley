# How many cluster are there?

Oftentimes we don't know how many clusters are in a dataset. One way to determine the number of cluster is to calulate the *"silhoutte score"* $\psi$ of a clustering (see Problem Set 2).<br>
Once the data points are labelled, one can measure the distance $d_i$ of a data point $x_i$ to its own cluster $S_i$ and the distance $d_j$ to the next closest cluster $S_j$. The idea is that for a good clustering $d_i < d_j$ and therefore<br>
<br>
$\psi(x_i) = \left\{ \begin{array}{ll}
    0, & \text{if } d_i = 0 \\
    \frac{d_j - d_i}{max(d_i, d_j)}, & \text{\, }
\end{array} \right.$<br>
<br>
We can do that for the entire dataset and return the average silhoutte score $\psi_{tot}$. The higher $\psi_{tot}$, the better the clustering.<br>
<br>
In this exercise we want to explore, how $\psi_{tot}$ depends on the structure of different cluster. In order to being able to visualize the cluster as they are, we want to keep it simple and start with 2D cluster first.
