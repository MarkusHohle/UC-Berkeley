**2) Preparation**

The motion of a particle (here, a bird) can be modelled as follows: the $x$ and $y$ coordinate (we want to stick to 2D) at time step $t+1$ can be derived from the previous time step $t$ by calculating Newton's equation of motion:<br>
<br>
$x(t+\Delta t) = x(t) + v_x(t)\, \Delta t + \frac{1}{2}a_x(t)\, \Delta t^2$<br>
$y(t+\Delta t) = y(t) + v_y(t)\, \Delta t + \frac{1}{2}a_y(t)\, \Delta t^2$<br>
<br>
$v_x(t+\Delta t) = v_x(t) + a_x(t)\, \Delta t$<br>
$v_y(t+\Delta t) = v_y(t) + a_y(t)\, \Delta t$<br>
<br>
where $\Delta t$ is the time step.<br>

The key point is to model the acceleration part $a_x$ and $a_y$. Acceleration is caused by a force $F$ via Newton's second law $F = m\,a$, so $a_x = \frac{F_x}{m}$ and $a_y = \frac{F_y}{m}$. More massive "birds" will react more slowly to changes. How do we model the force?<br> 
Birds tend to fly in flocks. That means that they don't want to be too far away from their friends. But on the other hand, they can't be arbitrarily close to eachother. Thus, there must be an ideal distance $r_{min}$. Such a behaviour can be modelled by a potential $\Phi(r)$ with a minimum at $r_{min}>0$ such as the Lennard-Jones potential (see below and Lecture Exercise).<br> 

<br>

We are not that far yet, but let us consider a force $F$, that is <br>
<br>
$F_x = -c\,x^2$<br>
$F_y = -c\,y^2$<br>

with a constant $c$.<br>
This force will cause birds that flew too far from the coordinate center (= the center of the flock) to being pulled back. At the coordinate center this force will be zero, but the bird will retain its momentum from the motion and therefore will fly out again, until the force pulls it back to the flock and so on. The simulation should show a "breathing" behaviour of the flock.<br>
Important: in contrast to a potential, a **force has a direction**! Make sure, that $F_x$ and $F_y$ point towards the center of the flock!
